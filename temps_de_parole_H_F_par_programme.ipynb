```python
import matplotlib.pyplot as plt 
import pandas as pd 
```

```python
df = pd.read_csv("data/ina-csa-parole-femmes-genreprogramme.csv") 
print(df.dtypes)
df.head()
```

```python
#j'utilise la fonction fillna() pour du coup remplacer les valeurs manquantes par 0, mais je peux choisir n'importe quel chiffre
df['men_speech_duration_2019'].fillna(0, inplace = True)
df['nb_declarations_2019'].fillna(0, inplace = True)
df['total_declarations_duration_2019'].fillna(0, inplace = True)
df['women_speech_duration_2019'].fillna(0, inplace = True)
#je peux aussi utiliser sans (inplace) comme ca : 
df['other_duration_2019'] = df['other_duration_2019'].fillna(0)
df['women_expression_rate_2019'] = df['women_expression_rate_2019'].fillna(0)
df['speech_rate_2019'] = df['speech_rate_2019'].fillna(0)
df.head(50)
```

```python
df1 = pd.DataFrame({
    'genre' : ['Information', 'Documentaire', 'Magazine', 'Divertissement', 'Sport', 'Musique'],
    'men_speech_2019' : [df['men_speech_duration_2019'].head(3).sum(), df['men_speech_duration_2019'].iloc[3], df['men_speech_duration_2019'].iloc[4], df['men_speech_duration_2019'].iloc[6:8].sum(), df['men_speech_duration_2019'].iloc[8], df['men_speech_duration_2019'].iloc[12]],
    'women_speech_2019' : [df['women_speech_duration_2019'].head(3).sum(), df['women_speech_duration_2019'].iloc[3], df['women_speech_duration_2019'].iloc[4], df['women_speech_duration_2019'].iloc[6:8].sum(), df['women_speech_duration_2019'].iloc[8], df['women_speech_duration_2019'].iloc[12]]
})
df1
#en fait avec juste iloc[4] comme ca, ca prend vrm du coup le numero 1 du tableau, dans dans notre caas cest magazine 
#alors que si tu fais iloc[6:8] ca prend du coup la position 5,6,8 en gos faut faire 6 5 1, avec iloc

df2 = pd.DataFrame({
    'genre' : ['Information', 'Documentaire', 'Magazine', 'Divertissement', 'Sport', 'Autres emissions', 'Non renseigné', 'Musique'],
    'men_speech_2020' : [df['men_speech_duration_2020'].head(3).sum(), df['men_speech_duration_2020'].iloc[3], df['men_speech_duration_2020'].iloc[4], df['men_speech_duration_2020'].iloc[6:8].sum(), df['men_speech_duration_2020'].iloc[8],df['men_speech_duration_2020'].iloc[10:11].sum(), df['men_speech_duration_2020'].iloc[11], df['men_speech_duration_2020'].iloc[12]],
    'women_speech_2020' : [df['women_speech_duration_2020'].head(3).sum(), df['women_speech_duration_2020'].iloc[3], df['women_speech_duration_2020'].iloc[4], df['women_speech_duration_2020'].iloc[6:8].sum(), df['women_speech_duration_2020'].iloc[8],df['women_speech_duration_2020'].iloc[10:11].sum(), df['women_speech_duration_2020'].iloc[11], df['women_speech_duration_2020'].iloc[12]]
})
df2
```

```python
plt.pie(df1['men_speech_2019'],  autopct= '%1.1f%%', colors= ['red', 'blue', 'green', 'yellow', 'darkviolet', 'orange', 'gold', 'silver'])
plt.title("Diagramme circulaire de temps de parole des hommes en fonction des programmes en 2019") 
plt.legend(labels= df1['genre'], bbox_to_anchor = (-0.2,0.75)) 
plt.show()
```

```python
plt.pie(df1['women_speech_2019'], autopct = '%1.1f%%', colors= ['red', 'blue', 'green', 'yellow', 'darkviolet', 'orange', 'gold', 'silver'])
plt.title("Diagramme circulaire de temps de parole des femmes en fonction des programmes en 2019") 
plt.legend(labels= df1['genre'], bbox_to_anchor = (-0.2,0.75)) 
plt.show()
```

```python
plt.pie(df2['men_speech_2020'], autopct = '%1.1f%%', colors= ['red', 'blue', 'green', 'yellow', 'darkviolet', 'orange', 'gold', 'silver'])
plt.title("Diagramme circulaire de temps de parole des hommes en fonction des programmes en 2020") 
plt.legend(labels= df2['genre'], bbox_to_anchor = (-0.2,0.75)) 
```
